```python
# Create an API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
def add_user_message(messages, text):
    user_message = {
        "role": "user",
        "content": text
    }
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {
        "role": "assistant",
        "content": text
    }
    messages.append(assistant_message)

def chat(messages):
    message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=messages
    )
    return message.content[0].text
```

```python
# Chat bot implementation

massages = []

while True:

    new_question = input("> ")
    print(">", new_question)

    add_user_message(massages, new_question)

    new_answer = chat(massages)
    
    print("---")
    print( new_answer)
    print("---")

    add_assistant_message(massages, new_answer)
```
